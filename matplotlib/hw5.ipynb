```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1, 5, 10, 15, 20]
y1 = [1, 7, 3, 5, 11]
y2 = [4, 3, 1, 8, 12]

plt.plot(x, y1, 'ro-', label='line 1')
plt.plot(x, y2, 'go-.', label='line 2')

plt.legend()
```

```python
plt.figure(figsize=(15, 7))

x = np.linspace(1, 5, 5)
y1 = [1.5, 6.5, 6, 3, 5]
y2 = [8.5, 4, 2, 4, 8.5]
y3 = [-7, -4, 2, -4, -7]

plt.subplot(2, 1, 1)
plt.plot(x, y1)

plt.subplot(2, 2, 3)
plt.plot(x, y2)

plt.subplot(2, 2, 4)
plt.plot(x, y3)
```

```python
x = np.linspace(-5, 5, 9)

plt.plot(x, x**2)

plt.annotate('min', xy = (0, 0), xytext=(0, 10), arrowprops=dict(facecolor='green'))
```

```python
data = np.random.randint(0, 11, size=(7, 7))

fig, ax = plt.subplots(figsize=(8, 6))
im = ax.imshow(data, cmap='viridis')

ax.set_xticks(np.arange(data.shape[1]))
ax.set_yticks(np.arange(data.shape[0]))

cbar = fig.colorbar(im, ax=ax, shrink=0.5)
```

```python
x = np.linspace(0, 5, 100)

y1 = np.cos(np.pi * x)

plt.plot(x, y1, 'r', linewidth=2)
plt.fill_between(x, y1, alpha=1)
```

```python
x = np.linspace(0, 5, 1000)
y = np.cos(np.pi * x)

y_masked = np.where(y < -0.5, np.nan, y)

#  
plt.plot(x, y_masked, linewidth=3)

plt.ylim(-1, 1)
```

```python
plt.figure(figsize=(20, 6))

x = np.arange(0, 7)
y = np.arange(0, 7)


plt.subplot(1, 3, 1)
plt.step(x, y, where='pre', color='g', marker='o')
plt.grid()

plt.subplot(1, 3, 2)
plt.step(x, y, where='post', color='g', marker='o')
plt.grid()

plt.subplot(1, 3, 3)
plt.step(x, y, where='mid', color='g', marker='o')
plt.grid()
```

```python
x = np.arange(0, 10, 0.1)
y1 = - x**2 / 5 + 2*x
y2 = -3*(x**2)/5 + 6*x
y3 = -2*(x**2)/3 + 26*x/3

plt.fill_between(x, y1, label='y1', zorder=3)
plt.fill_between(x, y2, label='y2', zorder=2)
plt.fill_between(x, y3, label='y3', zorder=1)

plt.legend(loc='upper left')

plt.ylim(0, 29)
plt.xlim(0, 10.5)
```

```python
labels = ['Ford', 'Toyota', 'BMW', 'AUDI', 'Jaguar']
sizes = [15, 10, 40, 12, 23]  #   

explode = (0, 0, 0.15, 0, 0)

plt.pie(sizes, labels=labels, explode=explode)

plt.axis('equal')
```
